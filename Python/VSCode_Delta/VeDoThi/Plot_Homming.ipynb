Đọc dữ liệu

```python
import matplotlib.pyplot as plt

plt.rcParams["savefig.directory"] = "C:/Downloads/"
rf = open("A:\Code\Python\VSCode_Delta\PhanTicFFT\INPUT_t.txt", "r")  ## Lưu ý: đổi lại đường dẫn tới file theo máy của bạn
read = rf.readlines()

# print(read)
del read[0:2]

t = []

x1 = []
y1 = []
z1 = []

x2 = []
y2 = []
z2 = []

x3 = []
y3 = []
z3 = []

for i in range(0, len(read)):
    if read[i] == "\t\t\t\t\t\t\t\t\t\n":
        break
    else:
        listRead = read[i].split("\t")
        t.append(float(listRead[0]))
        x1.append(float(listRead[1])-79+240)
        y1.append(float(listRead[2])-49+210)
        z1.append(float(listRead[3])-25+190)
        x2.append(float(listRead[4])-85+150)
        y2.append(float(listRead[5])-63+120)
        z2.append(float(listRead[6])-37+100)
        x3.append(float(listRead[7])-95+60)
        y3.append(float(listRead[8])-70+40)
        z3.append(float(listRead[9])-46+20)

# Number of sample points
N = len(t)
```

Cài đặt thông số

```python
linewidth = 2
fontsize = 28

x_min = 0
x_max = 35000
y_min = 0
y_max = 320
```

Plot Full1

```python
fig1 = plt.figure(figsize=(16, 8), dpi=120)

plt.plot(t, x1, label="x_1", color="blue", linewidth=linewidth, linestyle="dashed")
plt.plot(t, y1, label="y_1", color="red", linewidth=linewidth, linestyle="dashed")
plt.plot(t, z1, label="z_1", color="#00aa00ff", linewidth=linewidth, linestyle="dashed")

plt.title("Tín hiệu cảm biến số 1", fontsize=fontsize, weight="bold")
plt.xlabel("Thời gian (s)", fontsize=fontsize, weight="bold")
plt.ylabel("Tín hiệu (Analog)", fontsize=fontsize, weight="bold")

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)

plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)

plt.legend(fontsize=fontsize, loc="upper right")

plt.subplots_adjust(
    top=0.937, bottom=0.125, left=0.082, right=0.975, hspace=0.203, wspace=0.2
)

plt.grid()
plt.show()

# plt.savefig("figure1.png")
# plt.close(fig1)
```
