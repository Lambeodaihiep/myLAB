Đọc dữ liệu

```python
import matplotlib.pyplot as plt

plt.rcParams["savefig.directory"] = "D:/Downloads/"
rf = open("D:\Documents\VS_Code\Python\INPUT_Plot.txt", "r")  ## Lưu ý: đổi lại đường dẫn tới file theo máy của bạn
read = rf.readlines()

# print(read)
del read[0:3]

t1 = []
x1 = []
y1 = []
z1 = []

t2 = []
x2 = []
y2 = []
z2 = []



for i in range(0, len(read)):
  if read[i] == "\t\t\t\t\t\t\t\n":
    break
  else:
    listRead = read[i].split("\t")
    t1.append(float(listRead[0]))
    x1.append(float(listRead[1]))
    y1.append(float(listRead[2]))
    z1.append(float(listRead[3]))

    t2.append(float(listRead[4]))
    x2.append(float(listRead[5]))
    y2.append(float(listRead[6]))
    z2.append(float(listRead[7]))
    

# Number of sample points
N = len(t1)
```

Cài đặt thông số

```python
linewidth = 2
fontsize = 28

x_min = 0
x_max = 10
y_min = 0
y_max = 3
```

Plot Full1

```python
fig1 = plt.figure(figsize=(16, 8), dpi=120)

plt.plot(t1, x1, label="x_Old", color="blue", linewidth=linewidth, linestyle="dashed")
plt.plot(t1, y1, label="y_Old", color="red", linewidth=linewidth, linestyle="dashed")
plt.plot(t1, z1, label="z_Old", color="#00aa00ff", linewidth=linewidth, linestyle="dashed")

plt.title("Kết quả biến đổi Fourier tín hiệu\ncủa cảm biến số 3 khi chưa tăng khối lượng", fontsize=fontsize, weight="bold")
plt.xlabel("Tần số (Hz)", fontsize=fontsize, weight="bold")
plt.ylabel("Biên độ (Analog)", fontsize=fontsize, weight="bold")

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)

plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)

plt.legend(fontsize=fontsize, loc="upper right")

plt.subplots_adjust(
  top=0.937, bottom=0.125, left=0.082, right=0.975, hspace=0.203, wspace=0.2
)

plt.grid()
plt.show()

# plt.savefig("figure.png")
# plt.close(fig)
```
