Đọc dữ liệu

```python
import matplotlib.pyplot as plt
import gspread

plt.rcParams["savefig.directory"] = "D:/Downloads/"

sa = gspread.service_account(filename="A:\Code\Google_sheet\healthy-feat-390910-729c14c55127.json")

# lựa chọn 2 sheet cần so sánh
sh1 = sa.open("Dữ liệu rung Delta lần 22.1")
sh2 = sa.open("Dữ liệu rung Delta lần 22.2")

wks1 = sh1.worksheet("2. Động cơ quay với vi bước 1/32 - ko lọc") # tên worksheet muốn sử dụng, cái này mặc định rồi đừng có mà đổi
wks2 = sh2.worksheet("2. Động cơ quay với vi bước 1/32 - ko lọc")

data1 = wks1.get('A3:J')
data2 = wks2.get('A3:J')

# L là cột thời gian, M N O là cột cảm biến số 1, P Q R số 2, S T U số 3
# L bắt đầu từ chỉ số 0 của dữ liệu lấy được

t1 = []
x1 = []
y1 = []
z1 = []

t2 = []
x2 = []
y2 = []
z2 = []

# Number of sample points
N = len(data1)

for i in range(0, N):
    t1.append(float(data1[i][0]))

    x1.append(float(data1[i][1]))
    y1.append(float(data1[i][2]))
    z1.append(float(data1[i][3]))

    t2.append(float(data2[i][0]))

    x2.append(float(data2[i][1]))
    y2.append(float(data2[i][2]))
    z2.append(float(data2[i][3]))
```

Cài đặt thông số

```python
linewidth = 2
fontsize = 28

x_min = 0
x_max = 35000
y_min = 0
y_max = 140
```

So sánh X

```python
fig1 = plt.figure(figsize=(16, 8), dpi=120)

plt.plot(t1, x1, label="x_1", color="blue", linewidth=linewidth, linestyle="dashed")
plt.plot(t2, x2, label="x_2", color="red", linewidth=linewidth, linestyle="dashdot")

# plt.plot(t, y1, label="y_1", color="red", linewidth=linewidth, linestyle="dashed")
# plt.plot(t, z1, label="z_1", color="#00aa00ff", linewidth=linewidth, linestyle="dashed")

plt.annotate(text='', xy=(10000,10000), xytext=(0,0), arrowprops=dict(arrowstyle='<->'))

plt.title("Tín hiệu cảm biến số 1", fontsize=fontsize, weight="bold")
plt.xlabel("Thời gian (ms)", fontsize=fontsize, weight="bold")
plt.ylabel("Tín hiệu (Analog)", fontsize=fontsize, weight="bold")

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)

plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)

plt.legend(fontsize=fontsize, loc="upper right")

plt.subplots_adjust(
    top=0.937, bottom=0.125, left=0.082, right=0.975, hspace=0.203, wspace=0.2
)

plt.grid()
plt.show()
```
